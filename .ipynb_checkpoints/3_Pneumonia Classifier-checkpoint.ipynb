```python
batch_size = 32
img_height = 180
img_width = 180
img_folder='data\chest_xray'+folder
print(img_folder)
```

```python
train_ds = tf.keras.preprocessing.image_dataset_from_directory(
  img_folder,
  validation_split=0.2,
  subset="training",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)
```

```python
val_ds = tf.keras.preprocessing.image_dataset_from_directory(
  img_folder,
  validation_split=0.2,
  subset="validation",
  seed=123,
  image_size=(img_height, img_width),
  batch_size=batch_size)
```

```python
class_names = train_ds.class_names
print(class_names)
```

```python
plt.figure(figsize=(10, 10))
for images, labels in val_ds.take(1):
  for i in range(6):
    ax = plt.subplot(3, 3, i + 1)
    plt.imshow(images[i].numpy().astype("uint8"))
    plt.title(class_names[labels[i]])
    plt.axis("off")
```

```python
data_augmentation = keras.Sequential(
  [
    layers.experimental.preprocessing.RandomFlip("horizontal", 
                                                 input_shape=(img_height, 
                                                              img_width,
                                                              3)),
    layers.experimental.preprocessing.RandomRotation(0.1),
    layers.experimental.preprocessing.RandomZoom(0.1),
  ]
)
```

```python
num_classes = 2

model = Sequential([
  data_augmentation,
  layers.experimental.preprocessing.Rescaling(1./255),
  layers.Conv2D(16, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(32, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Conv2D(64, 3, padding='same', activation='relu'),
  layers.MaxPooling2D(),
  layers.Dropout(0.2),
  layers.Flatten(),
  layers.Dense(128, activation='relu'),
  layers.Dense(num_classes)
])
```

```python
model.compile(
  optimizer='adam',
  loss=tf.losses.SparseCategoricalCrossentropy(from_logits=True),
  metrics=['accuracy'])
```

```python
epochs=10
history=model.fit(
  train_ds,
  validation_data=val_ds,
  epochs=epochs
)
```

```python
his=pd.DataFrame(history.history)
his.plot()
```

```python
acc = history.history['accuracy']
val_acc = history.history['val_accuracy']

loss = history.history['loss']
val_loss = history.history['val_loss']

epochs_range = range(epochs)

plt.figure(figsize=(8, 8))
plt.subplot(1, 2, 1)
plt.plot(epochs_range, acc, label='Training Accuracy')
plt.plot(epochs_range, val_acc, label='Validation Accuracy')
plt.legend(loc='lower right')
plt.title('Training and Validation Accuracy')

plt.subplot(1, 2, 2)
plt.plot(epochs_range, loss, label='Training Loss')
plt.plot(epochs_range, val_loss, label='Validation Loss')
plt.legend(loc='upper right')
plt.title('Training and Validation Loss')
plt.show()
```

```python
# Classification Accuracy Measures

# Packages for classification accuracy
from sklearn.metrics import confusion_matrix
from sklearn.metrics import precision_recall_fscore_support
from sklearn.metrics import roc_curve, auc

import matplotlib.pyplot as plt
%matplotlib inline

# The following method determines and presents the following accuracy measures:
# Confusion Matrix
# Accuracy
# Precision
# Recall
# F1-score
# ROC
# AUC of ROC
def ClassificationAccuracy(actualTestOutcomes, predicted_probabilities, ClassThreshold = 0.5, title = ""):
    # Thresholded Accuracy Measures
    
    predicted_classes = 1*(predicted_probabilities > ClassThreshold)
    cm = confusion_matrix(actualTestOutcomes, predicted_classes)
    tn, fp, fn, tp = cm.ravel()
    accuracy = (tp + tn)/(tn + fp + fn + tp)
    precision, recall, fscore, support = precision_recall_fscore_support(actualTestOutcomes, predicted_classes)
    ConfusionPresent = (
        '                       Red  White\n' +
        'predict Red:     ' + str(tp) + "       " + str(fp) + "\n" + 
        'predict White:     ' + str(fn) + "       " + str(tn))
    accuracyMetrics = (
        "\nAccuracy: %0.6f" % accuracy +
        "\nPrecision: %0.6f" % precision[1] + 
        "\nRecall: %0.6f" % recall[1] + 
        "\nF1-score: %0.6f" % fscore[1])
    
    # ROC
    fpr, tpr, probability_thresholds = roc_curve(actualTestOutcomes, predicted_probabilities)
    AUC_metric = auc(fpr, tpr)

    plt.rcParams["figure.figsize"] = [6, 4]
    plt.figure()
    plt.title('Classification Accuracies  ' + title)
    plt.xlabel('FPR')
    plt.ylabel('TPR')
    plt.plot(fpr, tpr, LW=3, label='ROC curve (AUC = %0.6f)' % AUC_metric)
    plt.legend(loc="lower center")
    plt.plot([0, 1], [0, 1], color='navy', LW=3, linestyle='--') # reference line for random classifier
    #plt.text(x=0.1, y=0.0, s="AUC: %0.2f" % AUC_metric)
    plt.text(x=0.70, y=0.45, s=accuracyMetrics)
    plt.text(x=0.45, y=0.2, s=ConfusionPresent)
    plt.show()
    return None

# TrainAndTestWrapper combines model training and model testing
def TrainAndTestWrapper(model, X_train, y_train, X_test, y_test, title=""):
    model.fit(X_train, y_train) 
    test_probabilities = model.predict_proba(X_test)[:,1]
    ClassificationAccuracy(y_test, test_probabilities, title=title)
```

### Random Classifiers
A random classifier is a useful baseline model.  What accuracy do you expect from a random classifier on this balanced data set of 10 image types? 
- Create a function called ZeroModel that always returns 0.  
- Create a function called RandomModel that returns a random number between 0 and 9.

<br>Determine the accuracy of both of these "random" models. <span style="color:red" float:right>[2 point]</span>

### Assumptions
Assume that accuracy score from sklearn is adequate for this use.

```python
## your code goes here
def ZeroModel(image_array):
    out=[0]*len(image_array)
    return out
def RandomModel(image_array):
    from random import randrange
    out=[]
    for i in range(len(image_array)):
        out.append(randrange(10))
    return out
```

```python
from sklearn.metrics import accuracy_score
y_hat_train_Zero=ZeroModel(x_train)
y_hat_test_Zero=ZeroModel(x_test)
print('The accuracy score for a model that only returns 0 on the training data is',accuracy_score(y_train, y_hat_train_Zero))
print('The accuracy score for a model that only returns 0 on the test data is',accuracy_score(y_test, y_hat_test_Zero))
y_hat_train_Rand=RandomModel(x_train)
y_hat_test_Rand=RandomModel(x_test)
print('The accuracy score for a model that picks a random category on the training data is',accuracy_score(y_train, y_hat_train_Rand))
print('The accuracy score for a model that picks a random category on the test data is',accuracy_score(y_test, y_hat_test_Rand))
```

### Conclusion
The 0 only model has an accuracy score of 0.1 This makes sense as there are 60,000/10,000 images and both the training data set and the testing data set are balanced(i.e. there are the same amount of data for each label) So with 10 categories it makes sense that we get an accuracy of 0.1. The random model also returns a similar accuracy for the same reason(depending on the specific run it may be slightly higher or lower than 0.1).

### Define a neural net multiclass classifier
Using `keras`, define a neural net classifier, called `netmod`, to predict the label of the image.
1. Your model should first flatten each image. 
1. Then your model applies a dense layer of 5 neurons with a sigmoid activation.
1. Then your model applies a droput layer that drops 25 percent of neurons (dropout isn't really a layer, but the way it's implemented by `keras` makes it seem like a layer).
1. Then your model applies a dense layer of 10 for the multiclass classification of the 10 types of fashion items. This dense layer uses a 'sigmoid' activation function.

<br><span style="color:red" float:right>[5 point]</span>

NOTES ABOUT DROUPOUT LAYER

    keras.layers.Flatten(input_shape = (28, 28)),
    keras.layers.Dense(128, activation = tf.nn.relu),
    keras.layers.Dense(64, activation = tf.nn.relu),
    keras.layers.Dense(10, activation = tf.nn.softmax)
    tf.keras.layers.Dropout(rate, noise_shape=None, seed=None, **kwargs)

```python
## your code goes here
netmod = keras.Sequential([
    keras.layers.Flatten(input_shape = (28, 28)),# 1. flatten image
    keras.layers.Dense(5, activation = 'sigmoid'), # 2. 5 neurons with sigmoid
    keras.layers.Dropout(rate=0.25), # 3. dropout layer that drops(sets to 0) 25% of neurons of the above layer
    keras.layers.Dense(10, activation = 'sigmoid'), # 2. 10 neurons with sigmoid
])
```

### Compile the neural net
- for optimizer use `keras.optimizers.RMSprop` with a learning rate of 0.001
- for loss use `keras.losses.SparseCategoricalCrossentropy()`
- for metrics use `[keras.metrics.SparseCategoricalAccuracy()]`

<span style="color:red" float:right>[5 point]</span>

```python
## your code goes here
netmod.compile(optimizer = 'RMSprop', #same as above the default learning rate value is 0.001
              loss = 'sparse_categorical_crossentropy',
              metrics = 'sparse_categorical_accuracy')
```

### Train the neural net  
Train the model and assign the output to `training_history`
- 500 epochs
- batchsize of 2400 (divides an epoch into 25 batches) To help speed up training, we choose a larger number for `batch_size`. Note that in practice, the larger the `batch_size` the more mileage we get out of the array or tensor computations, but for large networks with millions of parameters, we eventually hit a limit because the resulting array exceeds available memory.

<br>Try out your training code with 2 epochs to make sure it runs before you use 500 epochs.  <span style="color:red" float:right>[5 point]</span>

```python
## your code goes here
training_history = netmod.fit(x_train, y_train, epochs = 50)
```

### Evaluate the training
Report the number of parameters in the model from `model.summary()`.  Use the values from `training_history.history` to plot the training accuracy of the model vs epoch number.  Compare the test accuracy with the training accuracy and determine if the model is overfit. <span style="color:red" float:right>[5 point]</span>
